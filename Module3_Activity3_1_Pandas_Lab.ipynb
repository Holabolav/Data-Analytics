```python
# Import required libraries
# Write code to import pandas and numpy
import numpy as np
import pandas as pd
```

## Task 1: Data Creation & Loading (1 Mark)
Dataset:
    'Name': ['Aman','Riya','Rahul','Sneha','Karan'],
    'Department': ['IT','HR','IT','Finance','HR'],
    'Salary': [50000,60000,55000,65000,52000]

```python
# Create:
# 1. A Pandas Series from a list of 5 elements
Series = pd.Series( ['Aman','Riya','Rahul','Sneha','Karan'])
print(Series)

print("\n")

# 2. A DataFrame with columns: Name, Department, Salary
Dataseries = {'Name': ['Aman','Riya','Rahul','Sneha','Karan'],
              'Department': ['IT','HR','IT','Finance','HR'],
              'Salary': [50000, 60000, 55000, 65000, 52000]}
df= pd.DataFrame(Dataseries)
print(df)

print("\n")
# 3. Write a dataset using (any one): CSV / Excel / JSON
df.to_csv('All_data.csv', index=False)
print("Database is saved as file name All_data.csv ")
```

## Task 2: Data Exploration (1 Mark)

```python
# Perform:
# Display first 3 rows
print(f"First three rows are {df.head(3)}")
print("\n")

# Display dataset information
print(f"Databse information: {df.info()}")
print("\n")

# Display statistical summary
print(f"statistical summary: {df.describe()}")
print("\n")
```

## Task 3: Indexing, Filtering & Sorting (1 Mark)

```python
# Perform:
# 1. Extract first 3 rows using .loc[]
print(f"First three rows: {df.loc[0:2]}\n")

# 2. Extract Salary column using .iloc[]
print(f"Salary column: {df.iloc[:, 2]}\n")

# 3. Filter employees with Salary > 55000
print(f"Employee with salary>55000:\n {df[df['Salary'] > 55000]}\n")

# 4. Sort dataset by Salary (descending)
sorted_df = df.sort_values(by='Salary', ascending=False)
print(f"Dataset sorted by Salary (descending):\n {sorted_df}")
```

## Task 4: Group-by & Aggregation (1 Mark)

```python
# Perform:
# 1. Find average salary department-wise
print(f"Average Salary Department-wise: {df.groupby('Department')['Salary'].mean()}\n")

# 2. Find total and count of employees per department
print(f"Total and count of employees per dept:\n {df.groupby('Department')['Salary'].agg(['sum', 'count'])}")
```

## Task 5: Data Cleaning & Time-Series (1 Mark)

```python
# Given:
df['Joining_Date'] = ['2022-01-10','2021-07-15','2023-03-12','2020-11-05','2022-05-20']

# Perform:
# 1. Convert Joining_Date to datetime
df['Joining_Date'] = pd.to_datetime(df['Joining_Date'])
print(f"After converting to datetime: \n{df}")


# 2. Extract Year and Month
df['Year'] = df['Joining_Date'].dt.year
df['Month'] = df['Joining_Date'].dt.month
print(f"\nAfter extracting Year and Month:\n {df}")

# 3. Identify duplicate rows
print(f"\nDuplicate Rows:\n {df[df.duplicated()]}")

# 4. Remove duplicate rows
df = df.drop_duplicates()
print(f"\nData after removing duplicates:\f {df}")
```
